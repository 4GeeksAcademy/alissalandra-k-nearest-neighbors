```python
import sqlite3
import pandas as pd

# URLs for the datasets
movies_url = "https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv"
credits_url = "https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv"

# Load the datasets into Pandas DataFrames
movies_df = pd.read_csv(movies_url)
credits_df = pd.read_csv(credits_url)

# Create a SQLite database connection
conn = sqlite3.connect('movies.db')

# Store DataFrames into separate tables in the database
movies_df.to_sql('movies', conn, index=False, if_exists='replace')
credits_df.to_sql('credits', conn, index=False, if_exists='replace')

# Perform SQL join operation to generate the third table containing unified information
query = """
    SELECT m.*, c.*
    FROM movies m
    JOIN credits c ON m.title = c.title
"""

# Execute the SQL query and store the result in a DataFrame
unified_df = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the first few rows of the unified DataFrame
print(unified_df.head())
```

```python
# Define the columns to keep
columns_to_keep = ['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']

# Select only the specified columns from the unified DataFrame
cleaned_df = unified_df[columns_to_keep]

# Display the first few rows of the cleaned DataFrame
print(cleaned_df.head())

```

```python
import json

# Extracting 'name' from JSON formatted columns and replacing original columns

# Extract and replace 'genres' column
cleaned_df['genres'] = cleaned_df['genres'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)]))

# Extract and replace 'keywords' column
cleaned_df['keywords'] = cleaned_df['keywords'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)]))

# Extract and replace 'cast' column (selecting only the first three names)
cleaned_df['cast'] = cleaned_df['cast'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)[:3]]))

# Display the transformed DataFrame
print(cleaned_df)
```

```python
# Extract and replace 'crew' column with director's name
cleaned_df['crew'] = cleaned_df['crew'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x) if item.get('job') == 'Director']))

# Convert 'overview' column to a list
cleaned_df['overview'] = cleaned_df['overview'].apply(lambda x: [x])

# Display the transformed DataFrame
print(cleaned_df)
```

```python
# Define the columns to process
columns_to_process = ['genres', 'cast', 'crew', 'keywords']

# Apply the function to remove spaces in each column
for column in columns_to_process:
    cleaned_df[column] = cleaned_df[column].str.replace(' ', '')

# Display the transformed DataFrame
print(cleaned_df)
```

```python
# Combine values from converted columns into 'tags' column
cleaned_df['tags'] = cleaned_df['genres'] + ',' + cleaned_df['keywords'] + ',' + cleaned_df['cast'] + ',' + cleaned_df['crew']

# Replace commas with blanks in 'tags' column
cleaned_df['tags'] = cleaned_df['tags'].str.replace(',', ' ')

# Drop the original converted columns
cleaned_df.drop(columns=['genres', 'keywords', 'cast', 'crew'], inplace=True)

# Display the transformed DataFrame
print(cleaned_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Vectorize the text data using TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the 'tags' column
tfidf_matrix = tfidf_vectorizer.fit_transform(cleaned_df['tags'])

# Display the shape of the TF-IDF matrix
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity
similarity = cosine_similarity(tfidf_matrix)

# Define the recommend function
def recommend(movie_index):
    # Calculate similarity scores
    distances = similarity[movie_index]
    
    # Get indices of top 5 most similar movies (excluding itself)
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Print the titles of top 5 similar movies
    print("Recommended movies for '{}':".format(cleaned_df.iloc[movie_index]['title']))
    for i in movie_list:
        print(cleaned_df.iloc[i[0]]['title'])

# Call the recommend function with a movie index
recommend(0)  # Example: recommend the movie at index 0

```
